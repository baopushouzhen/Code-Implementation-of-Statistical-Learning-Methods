```python
import numpy as np

class KdNode(object):
    def __init__(self, dom_elt, split, left, right):
        self.dom_elt = dom_elt  # 用来分隔的一个样本点，k维
        self.split = split  # 进行分割的维度，整数
        self.left = left  # 左平面。子树
        self.right = right  # 右平面，子树

class KdTree:
    def __init__(self, data):
        self.root = self.create_kdtree(data)
    
    def create_kdtree(self, data):
        k = data.shape[1]  # 数据维度
        
        def create_tree(split, data_set):
            if not data_set.any():  # 如果数据集为空
                return None
            
            order = np.argsort(data_set[:, split])
            data_set = data_set[order]
            
            split_position = len(data_set) // 2
            median = data_set[split_position]
            split_next = (split + 1) % k
            
            return KdNode(median, split, create_tree(split_next, data_set[:split_position]), create_tree(split_next, data_set[split_position + 1:]))
        
        return create_tree(0, data)
    
    def find_nearest(self, point):
        best = None
        best_dist = float('inf')
        
        def search_node(node, target, depth, best, best_dist):
            if node is None:
                return best, best_dist
            
            axis = node.split
            
            if target[axis] < node.dom_elt[axis]:
                next_node = node.left
                opposite_node = node.right
            else:
                next_node = node.right
                opposite_node = node.left
                
            best, best_dist = search_node(next_node, target, depth + 1, best, best_dist)
            
            if np.linalg.norm(target - node.dom_elt) < best_dist:
                best = node.dom_elt
                best_dist = np.linalg.norm(target - node.dom_elt)
                
            if np.abs(target[axis] - node.dom_elt[axis]) < best_dist:
                best, best_dist = search_node(opposite_node, target, depth + 1, best, best_dist)
                
            best, best_dist = search_node(next_node, target, depth + 1, best, best_dist)
            
            return best, best_dist
        
        return search_node(self.root, point, 0, best, best_dist)

# 示例用法
data = np.array([[2,3,9],[5,1,4],[9,4,6],[4,7,8],[8,8,1],[7,4,2]])
query_point = np.array([9,6,2])

kd_tree = KdTree(data)
nearest_point, nearest_dist = kd_tree.find_nearest(query_point)

print("Query Point:", query_point)
print("Nearest Point in KD Tree:", nearest_point)
print("Distance to Nearest Point:", nearest_dist)
```
